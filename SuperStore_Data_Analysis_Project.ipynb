```python
 
df1['Order Date'] = pd.to_datetime(df1['Order Date'])
df1['Year'] = df1['Order Date'].dt.year
yearly_sales = df1.groupby('Year')['Sales'].agg('sum').reset_index()
```

```python
#year over year
yearly_sales['Prev_Year_Sales'] = yearly_sales['Sales'].shift(1)
yearly_sales
```

```python
# Growth_Percentage
yearly_sales['YoY_Growth_%'] = (
    (yearly_sales['Sales'] - yearly_sales['Prev_Year_Sales']) 
    / yearly_sales['Prev_Year_Sales']
) * 100
yearly_sales
```
